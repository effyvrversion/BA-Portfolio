```python
#import libraries
import seaborn as sns
import pandas as pd
import matplotlib.pylab as plt
import datetime
from datetime import date
```

```python
#read file
#pd.set_option('display.max_rows', None)
df = pd.read_csv('games.csv')
df
```

Explore dataset: The orignal columns contain non-numeric data with multiple strings in a list, and date and numbers in 'object' format;

```python
#check columns;
df.columns
```

```python
#remove irrelevant columns;
df = df[[#'Unnamed: 0', 
       'Title', 'Release Date', 'Team', 'Rating','Times Listed',
       'Number of Reviews', 'Genres', 'Summary', 'Reviews', 'Plays',   
        'Playing', 'Backlogs', 'Wishlist']].copy()
df
```

```python
# heatmap to check the NA values in dataframe;
sns.heatmap(df.isna(),cbar = False)
# Missing values in both 'Team' column and 'Rating' column;
```

```python
df.isnull().sum()
```

```python
#locate the missing value in 'Team' column;
df[df['Team'].isna() == True]
```

```python
#update Na value;
df.at[1245, 'Team'] = "Hitsuji Goods"
```

```python
#locate the missing value in 'Rating' column;
df[df['Rating'].isna() == True]
```

```python
# fill Na values with 0.0 in the column;
df['Rating'] = df['Rating'].fillna(0.0)
```

```python
#locate the missing value in 'Summary' column;
df[df['Summary'].isna() == True]
```

```python
#replace Na with 'Unknown';
df.at[649, 'Summary'] = "Unknown"
```

```python
#replace "TBD" with 'now' date;
#convert date from object to a datetime format;

today = date.today()
datetime_now = today.strftime("%b %d, %Y")

df['Release Date'] = df['Release Date'].replace('releases on TBD', datetime_now)
df['Release Date'] = df['Release Date'].apply(lambda x: datetime.datetime.strptime(x,'%b %d, %Y'))
```

```python
# get the day from the date column
df['Day'] = df['Release Date'].dt.day
df['Month'] = df['Release Date'].dt.strftime('%b')
df['Year'] = df['Release Date'].dt.year
df['Week day'] = df['Release Date'].dt.day_name()
df
```

```python
#convert title, summery to str; 
df['Title'] = df['Title'].astype(str)
df['Summary'] = df['Summary'].astype(str)
```

```python
#convert the dtype of coulmns, which have the object as dtype
def rmv(x):
        y = x.split('K')
        return float(y[0])*1000 if len(y) == 2 else float(y[0])
```

```python
#convert Number of Reviews, Plays, Backlogs, Wishlist to numeric;
df['Number of Reviews'] = df['Number of Reviews'].apply(rmv)
df['Plays'] = df['Plays'].apply(rmv)
df['Backlogs'] = df['Backlogs'].apply(rmv)
df['Wishlist'] = df['Wishlist'].apply(rmv)
df['Times Listed'] = df['Times Listed'].apply(rmv)
df['Playing'] = df['Playing'].apply(rmv)
```

```python
df.dtypes
```

```python
# drop duplicates
df = df.drop_duplicates()
df
```

```python
#save file
df.to_csv('games_cleaned.csv')
```

```python
# clean the strings in the 'Team' columns;

df['Team'] = df['Team'].str.replace('[', '').str.replace(']', '').str.replace('\'', '').str.replace('"', '').str.split(', ')
df['Genres'] = df['Genres'].str.replace('[', '').str.replace(']', '').str.replace('\'', '')
df['Reviews'] = df['Reviews'].str.replace('[', '').str.replace(']', '').str.replace('\'', '').str.replace('"', '')
```

```python
# convert list elements of 'Team'to multiple rows;
# explode the'Team'and'Genres'columns;
df_team = df.explode('Team')
df_team
```

```python
# explode 'Genres'
df['Genres']=df['Genres'].str.split(', ')
df_genre = df.explode('Genres')
df_genre
```

```python
#locate empty values in Genres;
df_genre[df_genre['Genres'] == '']
```

```python
df_genre.to_csv('games_genre.csv')
```

```python
# popularity of high rating games;
import numpy as np

Title = top10_rating['Title']
Play = top10_rating['Plays']
Backlogs = 	top10_rating['Backlogs']
Wishlist = top10_rating['Wishlist']
barWidth = 0.25

plt.figure(figsize=(25, 10))
# Set position of bar on X axis
br1 = np.arange(len(Play))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

plt.bar(br1, Play, color ='#007ba7', width = barWidth,
        label ='Play')
plt.bar(br2, Backlogs, color ='#7ac5cd', width = barWidth,
         label ='Backlogs')
plt.bar(br3, Wishlist, color ='#9cbac7', width = barWidth,
         label ='Wishlist')

plt.xlabel("Top10 High-Rating Games")
plt.ylabel("Game Data")
plt.title("Top10 Top10 High-Rating Game Data")

plt.xticks([r + barWidth for r in range(len(Play))],
        ['Elden Ring','The Great Ace Attorney 2','Metal Gear Solid 3','Outer Wilds','Disco Elysium','Dwarf Fortress','Tokyo Necro','Half-Life: Alyx','Bloodborne','Sekiro: Shadows Die Twice' ],fontsize =10)
 
plt.legend()
plt.show()
```
